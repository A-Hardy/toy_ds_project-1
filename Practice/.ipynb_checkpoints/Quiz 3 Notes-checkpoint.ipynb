# CH.9 - Clustering

- Clustering involves separating a dataset into subgroups of related data
- Clustering is different from KNN/Regression in that it is __unsupervised__
    - we aren't making predictions (classifications) for a variable of which we already have observations
    - so no training the model
- We are using K-means algorithm
    - K refers to the number of clusters
    - Quality of a clister is the "within cluster sum of squared distances" (WSSD)
        - Squared distance from each point to the cluster centre
    - Goal is to minimize total WSSD for all clusters (sum of WSSDs)
- How K-means clustering works
    - Start with an initial clustering and then adjust iteratively until no more improvements can be made (min WSSD)
        1. Center update: move the center of each cluster
        2. Label update: update the cluster assignments of each observation 
        3. If WSSSD min, stop. Else repeat from step 1.
- What kind of data is suitable for K-means clustering  
     - Must be quantitative
- Use random restarts to prevent clusters from getting boxed out on the edge due to unlucky starting pos
    - Will appear optimized when actually a poor assignment
- Choosing number of clusters
    - The more clusters there are, the lower the WSSD will be
    - Choose the number of clusters where WSSD starts diminishingly decreasing
        - Elbow point
- Data must be standardized or the distance calculations for each cluster will be off
- Can do clustering on multiple variables, but more difficult to visualize

- New functions:
    - scale: standardizes variable
        - mutate(across(everything(), __scale__))
    - kmeans: initialize clustering
        - __kmeans__(standardized_data, centers = 3)
    - augment: takes kmeans data and original data, and combines them tidily
        - __augment__(penguin_clust, standardized_data)
        - tot.withinss: view WSSD of a cluster (not a function, but a column in the augmented dataset
            - glance(penguin_clust)
    - list: part of the rowwise + mutate to apply kmeans to the data
        - mutate(penguin_clusts = __list__(kmeans(standardized_data, k)))
    - pluck: extract element from the list
        - __pluck__(1)
    - unnest: turns elements in a column from a data-frame back into full data
        - __unnest__(glanced)
    - nstart: initializes with multiple random cluster starting location
        - mutate(penguin_clusts = list(kmeans(standardized_data, __nstart = 10__, k))
    - ggpairs: visualizing relationship between variables
        - ggpairs(dataset)
        

```r
#needed for clustering functions
library(tidyverse)
library(broom) #needed for augment function

options(repr.matrix.max.rows = 6)
```

### Clustering for 2 variables

```r
#reproducibility
set.seed(17)

#reading in and cleaning the data
pm_data <- read_csv('data//pokemon.csv')|>   #read in the data
    select(Speed, Defense) |>   #select desired columns
    map_df(scale)   # scale all predictors

#applying clustering
cluster_ks <- tibble(k = 1:10) |>   #create tibble with the k-values to test
    rowwise() |>
    mutate(cluster_ks = list(kmeans(pm_data, nstart = 10, k)),  #initialize 10 times
              glanced = list(glance(cluster_ks))) |>
    unnest(glanced)

#view the elbow plot
elbow_plot <- ggplot(cluster_ks, aes(x = k, y = tot.withinss)) +
  geom_point() +
  geom_line() +
  xlab("K") +
  ylab("Total within-cluster sum of squares") +
  scale_x_continuous(breaks = 1:9) + 
  theme(text = element_text(size = 12))
elbow_plot
```

```r
# use found optimal number of clusters
pokemon_final_kmeans <- kmeans(pm_data, nstart = 10, centers = 3) #found k = 3
pokemon_final_clusters <- augment(pokemon_final_kmeans, pm_data) 
    
#view clustered plot
pokemon_final_clusters_plot <- ggplot(pokemon_final_clusters, aes(x = Speed, y = Defense, color = .cluster)) +
    geom_point()+
    labs(x = "Pokemon Speed", y = 'Pokemon Defence', color = "Cluster")
pokemon_final_clusters_plot
```

### Clustering for multiple variables

```r
set.seed(2019)
#reading in the data
clean_reviews <- read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00484/tripadvisor_review.csv') |>
    select(-"User ID")

#apply clustering
cluster_reviews <- tibble(k = 1:10) |>  #trying k-values 1-10
    rowwise() |>
    mutate(cluster_reviews = list(kmeans(clean_reviews, nstart = 100, k)), #starting with 100 initializations
         glanced = list(glance(cluster_reviews)))

#get stats
elbow_stats <- cluster_reviews |>
      select(- cluster_reviews) |>
      unnest(glanced)

#generate plot
tourism_elbow_plot <- ggplot(elbow_stats, aes(x = k, y = tot.withinss)) +
  geom_point() +
  geom_line() +
  xlab("K") +
  ylab("Total within-cluster sum of squares") +
  scale_x_continuous(breaks = 1:9) + 
  theme(text = element_text(size = 12))
tourism_elbow_plot
```

```r
set.seed(2019)

reviews_clusters <- kmeans(clean_reviews, nstart = 100, centers = 2) #chose k = 2
reviews_clusters #look at cluster means to see the average for each variable in each cluster

cluster_assignments <- augment(reviews_clusters, clean_reviews) #to view dataset with assigned clusters as a column
cluster_assignments
```

```r
#viewing distribution of variables. Variables with different means should drive the clustering ie cat 10, 3, 5, 6, 7.
options(repr.plot.height = 8, repr.plot.width = 15)
cluster_assignments |>
    pivot_longer(cols = -.cluster, names_to = 'category', values_to = 'value')  |> 
    ggplot(aes(value, fill = .cluster)) +
        geom_density(alpha = 0.4, colour = 'white') +
        facet_wrap(~ category, scales = 'free') +
        theme_minimal() +
        theme(text = element_text(size = 20))
```

# CH.10 - Statistical Inference

- We typically don't get data on the entire population, only a sample. Need to infer things from the sample about the population. 
    - Will never really know exactly how accurate our inference is without access to true population parameter
- Can be used on categorical parameter (and estimate proportion of that category)
- Can be used on a quantitative parameter (and estimate mean)
- Population and sample may have similar shape, but sampling distribution will likely not (should be ~normally distributed if size is big enough)
    - ie. if population distribution is skewed, the sample you take should likewise be skewed
        - the sampling distribution, however will be normally distributed if sample size is sufficient
    - sampling distribution is an average, within that there will be some higher and some lower values so the mean will never become too extreme.
- Improve the point estimate by taking a larger sample
    - Sampling distribution will get taller and narrower, with a more defined peak closer to the mean
<br> .
    
- Bootsrapping:
    1. Randomly sample WITH replacement until sample size n (you decide). This makes up one bootstrap.
        - Boostrap size can be the same as size of sample, but will not necessarily include all values in the sample b/c you sample with replacement.
    2. Calculate the point estimate from those samples (ie. mean, median...)
    3. Repeat with the desired number of bootstraps.
    4. Plot the point estimates of each boostrap to create the boostrap distribution. 
        - The boostrap distribution (sampled from a sample) shape should resemble the sampling distribution (sampled from the population).
        - The mean of boostrap distribution will not necessarily be centered around the same mean as the sampling distribution. 
            - Boostrap dist will be centered around the parameter of the sample you sampled from. Sampling dist will be centered around population parameter. 
    - Bootstraps will differ from eachother because you sample with replacement
    - 95% confidence interval can be taken from the bootstrap distribution
        - Exclude bottom and top 2.5% of values
        - You can be 95% sure that the true mean falls somewhere in that range

- Definitions
     - __population__: the entire collection of cases we're interested in (usually entire population is inaccessible)
     - __population parameter__: numerical characterization of the entire population
     - __sample__: subset of individuals from the population
     - __sample estimate__: numerical characteristic of the sample (an estimate of the population parameter)
     - __statistical inference__: using the sample to make a conclusion about the entire population
     - __point estimate__: single value estimated from a sample
     - __sampling variability__: variation from sample to sample
     - __sampling distribution__: distribution of many estimates for all available samples from the population. Centered at the estimated pop. mean.
     - __interval estimation__: 
     - __confidence interval__: range of plausible values for the population parameter
     - __bootstrap__: assuming our sample is big enough, it's distribution should resemble the population. We can then take many sub-samples from this sample to mimic taking samples from the population (since the whole population is inaccessible). 
     - __boostrap distribution__: distribution of sample means taken from the boostrap sample (not the population). 
     
- New functions
    - __rep_sample_n__: randomly select samples from the population dataset.
        - rep_sample_n(tbl = population_dataset, size = 40, reps = 20000) 
            - size is the size of each sample to take
            - reps is the number of samples to take
    

```r
#library(tidyverse) loaded above for clustering so won't load here again, but would need normally
library(repr)
library(digest)
library(infer)
```

```r
#simulate a population to work with
set.seed(17) 
students_pop <- tibble(grade = (rnorm(mean = 70, sd = 8, n = 10000))) #make a sample of 10,000 students, mean grade = ~70% (randomized so not perfect)
students_pop

#get population parameters (just to reference, in a real setting wouldn't actually have access to this)
pop_parameters <- students_pop  %>% 
    summarize(pop_mean = mean(grade), pop_med = median(grade), pop_sd = sd(grade))
pop_parameters
```

```r
# how to draw one sample from population:
set.seed(17)
ests_5 <- rep_sample_n(students_pop, size = 5)  %>% #draw one sample of 5 students
    summarize(mean_5 = mean(grade), med_5 = median(grade), sd_5 = sd(grade)) #view summary stats of sample
ests_5
```

```r
# create sampling distribution
set.seed(4321) 
sampling_distribution_20 <- rep_sample_n(students_pop, size = 20, reps = 1500) |> #lots of samples needed for sampling distribution
    group_by(replicate)  %>% 
    summarize(sample_mean = mean(grade)) |>
    ggplot(aes (x = sample_mean)) +
        geom_histogram(binwidth = 0.5) + #lower this is bars look too fat
        labs(title = "Distribution of sampled grades", x = "Average grade n=5")
sampling_distribution_20

#increasing sample size will make figure narrower and more centered around the mean
```

```r
#bootstrapping 

#making dataset
can_seniors <- tibble(age = (rexp(2000000, rate = 0.1)^2) + 65) |>  #right skewed dist
    filter(age <= 117, age >= 65)
#can_seniors

#take one sample, simulates sampling from the population IRL
one_sample <- can_seniors |> 
    rep_sample_n(40) |> 
    ungroup() |> # ungroup the data frame 
    select(age) # drop the replicate column 
#one_sample

#boostrapping
boot_est_means <- rep_sample_n(one_sample, size = 40, replace = TRUE, reps = 1000) |> #taking 1000 bootstraps
    group_by(replicate) |>
    summarize(mean = mean(age))
#boot_est_means

#boostrap distribution
boot_est_dist <- boot_est_means |>
    ggplot(aes(x = mean)) +
        geom_histogram(binwidth = 1) +
        labs(title = "1000 Boostraps", x = "Mean age")
boot_est_dist

#95% confidence interval 
boot_est_means |> 
    select(mean) |> 
    pull() |> 
    quantile(c(0.025, 0.975)) #change these if you wnt different from 95% CI
```
